- Cleaning the Data

```python
#Adding a new column 'Gender' to the dataframe
df['Gender']=4
df['Gender'].head(3)
```

```python
# mapping female, male into integers
df['Gender'] = df['Sex'].map({'female':0,'male':1}).astype(int)
df['Gender'].head(3)
```

```python
# Dealing with null values by taking the median as their values
#1.Create an alternative column for age
df['AgeFill']=df['Age']
df[df['Age'].isnull()][['Gender','Pclass','Age','AgeFill']].head(10)
```

```python
#2.
# Taking the median for each gender and pclass
med_ages = np.zeros((2,3))
for i in range(2):
    for j in range(3):
      med_ages[i,j]=df[(df.Gender==i)&(df.Pclass==j+1)]['Age'].\
    dropna().median()
for i in range(2):
    for j in range(3):
        df.loc[(df.Gender==i)&(df.Pclass==j+1)&(df['Age'].isnull()),\
               'AgeFill']=med_ages[i,j]
df[df['Age'].isnull()][['Gender','Pclass','Age','AgeFill']].head(10)
```

```python
# Creating a column that tells us whether the Age column 
# was originally null or not
df['AgeIsNull']=pd.isnull(df.Age).astype(int)
df.head(1)
```
